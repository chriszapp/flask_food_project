Using the library [openfoodfacts](https://pypi.org/project/openfoodfacts/) to download data and saving it to openfoodfacts_products.csv

```python
import pandas as pd
import numpy as np
import openfoodfacts
```

```python
# Grabbing brands
prod_frombrand = pd.DataFrame(openfoodfacts.facets.get_brands())[['id', 'products']]\
    .rename(columns = {'products': 'products_frombrand'}).astype({'id': str, 'products_frombrand': int})
print('prod_frombrand shape:', prod_frombrand.shape)
prod_frombrand.head()
```

```python
# Grabbing products
prod_fromstore = pd.DataFrame(openfoodfacts.facets.get_stores())[['id', 'products']]\
    .rename(columns = {'products': 'products_fromstore'}).astype({'id': str, 'products_fromstore': int})
print('prod_fromstore shape:', prod_fromstore.shape)
prod_fromstore.head()
```

```python
# Mergin tables on id
products = prod_fromstore.merge(prod_frombrand, how = 'outer', on = 'id').fillna(0).astype({'products_fromstore': int, 'products_frombrand': int})
products.head()
```

```python
# Save data in csv
products.iloc[:1000].to_csv('openfoodfacts_products.csv', index = False)
```
